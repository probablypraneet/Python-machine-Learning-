```python
# making cars Dictionary

Car_Brands = {"Chevrolet":"Corvett","Mitsubishi":"Lancer Evo", "Mercedes_Benz":"SL300"}
#3 of my favourite car brands from each country 
# here the key variables are the CAR BRANDS itself and the values are their Model names


print("Dream car: ",Car_Brands["Mitsubishi"]) # accessing the value of Mitsubishi car Brand 

# adding my favcourite car model form Ford and deleting the least wanted 

Car_Brands["Ford"]= "GT40"#added
del Car_Brands["Chevrolet"]#removed

for key,value in Car_Brands.items():
    print(key,": ",value)
```

```python
#Making Dictionary for the third time 

Car_Country = {"Ford":"America", "Mitsubishi":"Japan","Mercedes-Benz":"German"}

print("the favourite Car form Japan", Car_Country.items())

# add and remove 

Car_Country["Aston Martin"]= "British" #Add
print("the favourite Car form Japan", Car_Country.items())

del Car_Country["Ford"] #remove 

for key,value in Car_Country.items():
    print(key,value)
```
