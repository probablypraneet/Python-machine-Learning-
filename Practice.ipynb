```python
# operators 

# addition operators

y = 5
y+=4
print(f'5+4 = {y}')

#subtraction operator 

a= 20
a-=10
print(f'20 -10 = {a}')

# multiplication

b = 30
b*=2
print(f'30*2= {b}')

# division

d = 23
d/= 3
print(f'23 div by 3: rounded off ans is {d:.2f}')

#floor division 
d1=23
d1//=3
print(f'23 floor div by 3 ans is {d1}')


#exponent / power

p1 = 22
p1**=2

print(f'22^2 = {p1}')

#modulus op

m = 33
m%= 2

print(f'the remainder of 33/2 is {m} and the answer is {33/2}')






```

```python
#exception handling:


def intInput():

    try:
        nm1 = int(input("Enter a number"))
        nm2 = int(input("Enter number 2"))
    
        r = nm1 +nm2
        print(f"{nm1} + {nm2}  = {r}")

    except ValueError:
        print("ERROR::>>!!Enter Valid Integer Number only!!")

intInput()
```
